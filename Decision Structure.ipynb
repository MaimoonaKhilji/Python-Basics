## Conditional Expression (Boolean Expression)

Python provides various operations for comparing values. The result of a comparison is a Boolean Value, either **True** or **False**.

```python
#For example

2 > 10         #2 is not greater than 10, so result will be False
```

```python
 2 < 10      #As 10 is greater than 2, so result will be True
```

## Relational Operator (Comparison Operator)

These are used to compare values. It either returns True or False according to the condition.

```python
#Lets try few examples

x = 5
y = 7

print( 'x > y is ', x > y)

print( 'x==y is ', x==y)

print( 'x!=y is ',x!=y )
```

## Logical Operators

Logical Operators are the **and, or, not** operators. Each of them operates on one or more bool type and returns a bool type result. They ate used to check more than one condition.

```python
#Lets try few examples


print("2==2 and 2<10", 2==2 and 2<10)
```

```python
print("2!=2 or 2<10", 2!=2 or 2<10) # here 2<10 is true so result will be true
```

```python
print ( not 2==2) #not complements the result. Here the 2==2 gives true but because of not operand, it becomes False 
```

## If-Else Statement

When a user has more than one condition or there is possibility that an event can occur or not.

```python
#Example

x = 2
if x==2:
    print(x, " is equal to 2")
else:
    print(x, " is not equal to 2")
```

```python
x = 4
if x==2:
    print(x, " is equal to 2")
else:
    print(x, " is not equal to 2")
```

## If-Elif-Else Statements


If there are more than 2 choices, then we use elif command also. ELif is just if command that needs a condition.

```python
#For example

x = 5

if x < 10:
    print("x is smaller than 10")
    
elif x > 10:
    print("x is larger than 10")
```

```python
x = 15

if x < 10:
    print("x is smaller than 10")
    
elif x > 10:
    print("x is larger than 10")
```
